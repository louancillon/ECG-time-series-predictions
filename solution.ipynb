```python
#Create a list of Series with each signal
ecg_signals = [data.iloc[i] for i in range(data.shape[0])]
#clean the signals
cleans = [nk.ecg_clean(ecg_signals[i], sampling_rate=300, method="biosppy") for i in range(data.shape[0])]
```

```python
#Find the position of the Rpeaks for each signal
rpeaks = [nk.ecg_peaks(cleans[i], sampling_rate=300)[1]['ECG_R_Peaks'] for i in range(data.shape[0])]
```

```python
#Extract S, Q, P, T peaks 
#Bcp trop long mais ca pourrait etre bien utile ! 
#TODO trouver un autre moyen 
#waves_peaks = [nk.ecg_delineate(cleans[i], rpeaks[i], sampling_rate=300, method="peak")[1] for i in range(data.shape[0])] 
```

```python
#Values of the signal at the Rpeaks points 
rpeaks_values = [[cleans[i][rpeaks[i][j]] for j in range(len(rpeaks[i]))] for i in range(data.shape[0])]  
```

```python
#Calculate the min, max, mean and std values of the rpeaks for each signals
rpeaks_mean = [statistics.mean(rpeaks_values[i]) for i in range(data.shape[0])]
rpeaks_median = [statistics.median(rpeaks_values[i]) for i in range(data.shape[0])]
rpeaks_min = [min(rpeaks_values[i]) for i in range(data.shape[0])]
rpeaks_max = [max(rpeaks_values[i]) for i in range(data.shape[0])]
rpeaks_std = [statistics.stdev(rpeaks_values[i]) for i in range(data.shape[0])]
```

```python
#Calculate RR intervals 
rr_intervals = [[rpeaks[i][j+1] - rpeaks[i][j] for j in range(len(rpeaks[i])-1)] for i in range(data.shape[0])]
rr_mean = [statistics.mean(rr_intervals[i]) for i in range(data.shape[0])]
rr_median = [statistics.median(rr_intervals[i]) for i in range(data.shape[0])]
rr_min = [min(rr_intervals[i]) for i in range(data.shape[0])]
rr_max = [max(rr_intervals[i]) for i in range(data.shape[0])]
rr_std = [statistics.stdev(rr_intervals[i]) for i in range(data.shape[0])]
```

```python
#heartbeats signals 
#TODO qu'est ce qu'on peut en extraire comme info utiles 
heartbeats = [ecg.extract_heartbeats(signal=cleans[i], rpeaks=rpeaks[i], sampling_rate=300)[0] for i in range(data.shape[0])]
```

```python
## Group all the features together 
#J'ai gardé que les stats pas les arrays
#Je sais pas trop comment les utiliser apres pour les predictions
'''
datalist = list(zip(rpeaks,rpeaks_values, rr_intervals, heartbeats, 
               rpeaks_mean, rpeaks_min, rpeaks_max, rpeaks_median, rpeaks_std,
               rr_mean, rr_min, rr_max, rr_median, rr_std))
df = pd.DataFrame(datalist,  columns=['rpeaks','rpeaks_values', 'rr_intervals', 'heartbeats', 
               'rpeaks_mean', 'rpeaks_min', 'rpeaks_max','rpeaks_median', 'rpeaks_std',
               'rr_mean', 'rr_min', 'rr_max', 'rr_median', 'rr_std'])
               '''
datalist = list(zip(rpeaks_mean, rpeaks_min, rpeaks_max, rpeaks_median, rpeaks_std,
               rr_mean, rr_min, rr_max, rr_median, rr_std))
df = pd.DataFrame(datalist,  columns=['rpeaks_mean', 'rpeaks_min', 'rpeaks_max','rpeaks_median', 'rpeaks_std',
               'rr_mean', 'rr_min', 'rr_max', 'rr_median', 'rr_std'])
print(df)
```

```python
#Feature selection 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesClassifier

# Building the model
extra_tree_forest = ExtraTreesClassifier()
  
# Training the model
extra_tree_forest.fit(df, y_train.values.ravel())
  
# Computing the importance of each feature
feature_importance = extra_tree_forest.feature_importances_
  
# Normalizing the individual importances
feature_importance_normalized = np.std([tree.feature_importances_ for tree in 
                                        extra_tree_forest.estimators_],
                                        axis = 0)
```

```python
# Plotting a Bar Graph to compare the models
plt.bar(df.columns, feature_importance_normalized)
plt.xlabel('Feature Labels')
plt.ylabel('Feature Importances')
plt.title('Comparison of different Feature Importances')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


test = df[['rr_mean', 'rr_min', 'rr_max', 'rr_median', 'rr_std']]
x_train, x_test, y_trainn, y_test = train_test_split(test, y_train, test_size=0.20)
# Use the KNN classifier to fit data:
classifier = KNeighborsClassifier(n_neighbors=4)
classifier.fit(x_train, y_trainn) 

# Predict y data with classifier: 
y_predict = classifier.predict(x_test)
```

```python
from sklearn.metrics import f1_score

F1 = f1_score(y_test, y_predict, average='micro')
print(F1)
```
